# Baseball Salaries

In this project, I will be analyzing a baseball dataset that tracks salaries of players from the National League and the American League from 1985 to 2016.

The data can be found at https://www.seanlahman.com/baseball-archive/statistics/

It want to find out who the highest and lowest earners are and how the salaries increased over the years.
    


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Salaries.csv')
```

```python
print(data.describe())
```

```python
data[['yearID', 'playerID', 'salary']]
```

```python
total_salary_year = data.groupby('yearID')['salary'].agg(sum)
total_salary_year.head(32)
```

```python
total_salary_year.plot(figsize=(15,10))
plt.axis([1985,2016, 0, 4000000000])
plt.title("Sum of all players salaries in both AL and NL by year")
plt.xlabel('Year')
plt.ylabel('Total league salary (billions)')
```

```python
ts = data.groupby(['lgID', 'yearID'])['salary'].agg(sum)
ts = ts.reset_index(level="lgID")
salary_pt =  pd.pivot_table(ts, values='salary', index=["yearID"], columns=["lgID"])
salary_pt
```

```python
salary_pt.plot(figsize=(15, 12))
```
